```python
#Leitura de todos os arquivos da pasta song_data e a concatenacao para a criacao de um dicionario 
path_to_jsonfiles = 'C:/Users/55199/Desktop/CV/Iblue/data/song_data/A/A/A'
alldicts = []
for file in os.listdir(path_to_jsonfiles):
    full_filename = "%s/%s" % (path_to_jsonfiles, file)
    with open(full_filename,'r') as fi:
        dict = json.load(fi)
        alldicts.append(dict)
```

```python
alldicts
```

```python
##convertendo um json dataset de um dicionario para um dataframe
df5 = pd.DataFrame(alldicts)
```

```python
df5.columns
```
